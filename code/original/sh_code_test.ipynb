```python
methods = ['knn', 'knn_std', 'mf', 'mf_std', 'p1', 'p1_std', 'p2', 'p2_std', 'p3', 'p3_std']
data_split = ['train', 'val']

for method in methods:
    for split in data_split:
        for data_type in ['X_', 'y_']:
            for co in ['uncorr_', 'corr_']:
                for thresh in ['95', '100']:
                    print(method)
                    print(data_type + split + '_' + co + thresh + '.parquet')
```
